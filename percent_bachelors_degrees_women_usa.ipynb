This project aims to present clear and objective visualizations to support governmental efforts in developing campaigns to promote the advancement of women in universities and the workforce. By understanding the gender gap in bachelor's degree attainment, the project provides valuable insights that can help design effective strategies for encouraging female participation and success in higher education and career fields.

[The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp) releases a data set annually 
containing the percentage of bachelor's degrees granted to women in the U.S.A. from 1970 to 2012. The data set is broken up 
into 17 categories of degrees, with each column as a separate category.

[Randal Olson](http://www.randalolson.com), a data scientist at University of Pennsylvania, has cleaned the data set and made 
it available on his personal website. You can download the dataset Randal 
compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). 
You can see his approach [here](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/).

In this project we use data visualization to show the gender gap in all the fields.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent_bachelors_degrees_women_usa.csv')
```

```python
women_degrees.head()
```

```python
# Colors
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
# Major categories
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 
             'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 
              'Architecture']
```

```python
fig = plt.figure(figsize=(15, 18))
st = fig.suptitle('Bachelor Degree Gender Gap from 1970 to 2012', fontsize='xx-large')

# shift title down:
st.set_y(0.93)

# plotting stem_cats column
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3)+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(stem_cats[sp])
    # removing the spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # set x,y range
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # remove ticks
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    # add a horizontal line at half
    ax.axhline(50, c='gray', alpha=0.3)
    # specifying which labels we want displayed
    ax.set_yticks([0,100])
    # adding text anotation for the lines
    if sp == 0:
        ax.text(2003, 80, 'Women')
        ax.text(2004, 14, 'Men')
    elif sp == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 10, 'Women')
        ax.tick_params(labelbottom=True)
     

# plotting lib_arts_cats column
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3)+2)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, linewidth=3)
    ax.set_title(other_cats[sp])
    # removing the spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # set x,y range
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # remove ticks
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    # add a horizontal line at half
    ax.axhline(50, c='gray', alpha=0.3)
    # specifying which labels we want displayed
    ax.set_yticks([0,100])
    # adding text anotation for the lines
    if sp == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2004, 7, 'Men')
    elif sp == 5:
        ax.text(2005, 60, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(labelbottom=True)

# plotting lib_arts_cats column    
for sp in range(0,5):
    ax = fig.add_subplot(6,3,(sp*3)+3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, linewidth=3)
    ax.set_title(lib_arts_cats[sp])
    # removing the spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # set x,y range
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # remove ticks
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    # add a horizontal line at half
    ax.axhline(50, c='gray', alpha=0.3)
    # specifying which labels we want displayed
    ax.set_yticks([0,100])
    # adding text anotation for the lines
    if sp == 0:
        ax.text(2003, 75, 'Women')
        ax.text(2004, 20, 'Men')
    # if sp == 5:
    #     ax.tick_params(labelbottom=True)

# adding the x-axis lables
ax = fig.add_subplot(6,3,18)
# removing the spines
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)
# set x,y range
ax.set_xlim(1968, 2011)
ax.set_ylim(0,100)
# remove ticks
ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
# remove y-axis
ax.set_yticks([])
# adding x-axis labels
ax.tick_params(labelbottom=True)

# plt.savefig('gender_gap.jpg')
plt.show()
```

Engineering and computer science face challenges in balancing the representation of women in these fields.

We can also see that:

- Fields like Physical Sciences, Agriculture, Social Sciences, History, Business, Engineering, and Architecture have been getting better at closing the gender gap over time.
- On the other hand, Psychology, Health Professions, Public Administration, and Computer Science have been seeing a bigger gender gap.
- In Communication and Journalism, the percentage of women has grown over time and now exceeds that of men, with a major shift happening around 1980.
- The gender gap in Engineering has slowly decreased over theyears.
